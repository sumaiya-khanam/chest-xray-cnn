# Imports Library

```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

# Dataset Paths 

```python
lung_opacity_path = "/kaggle/input/lung-disease/Lung X-Ray Image/Lung X-Ray Image/Lung_Opacity"
normal_path = "/kaggle/input/lung-disease/Lung X-Ray Image/Lung X-Ray Image/Normal"
viral_pneumonia_path = "/kaggle/input/lung-disease/Lung X-Ray Image/Lung X-Ray Image/Viral Pneumonia"
```

# Function to load Sample Images and Plotting

```python
def load_samples(path, num_samples=3): 
    files = os.listdir(path)[:num_samples]
    images = [Image.open(os.path.join(path, f)) for f in files]
    return images

lung_opacity_imgs = load_samples(lung_opacity_path)
normal_imgs = load_samples(normal_path)
viral_pneumonia_imgs = load_samples(viral_pneumonia_path)

fig, axes = plt.subplots(3, 3, figsize=(10, 10))  


for i, img in enumerate(lung_opacity_imgs):
    axes[0, i].imshow(img, cmap='gray')
    axes[0, i].axis('off')
axes[0, 1].set_title("Lung Opacity", fontsize=16, fontweight='bold') 


for i, img in enumerate(normal_imgs):
    axes[1, i].imshow(img, cmap='gray')
    axes[1, i].axis('off')
axes[1, 1].set_title("Normal", fontsize=16, fontweight='bold')


for i, img in enumerate(viral_pneumonia_imgs):
    axes[2, i].imshow(img, cmap='gray')
    axes[2, i].axis('off')
axes[2, 1].set_title("Viral Pneumonia", fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
